# Ejemplo: Aproximación local alrededor de $x_0$ con series de Taylor

Aproximar la función $f(x) = x^4$, alrededor de $x_0=4$ con series de Taylor

$$f(x)$$

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def f(x):
    return x**4

def df(x):
    return 4*x**3
    
def ddf(x):
    return 12*x**2
    
def taylor(x,x0=4):
    d = x-x0
    return f(x0)+d*df(x0)+0.5*d*ddf(x0)*d
```

```python
x = np.arange(-4,4,0.1)
y = [f(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
xc=2

local_range = 1
x = np.arange(xc-local_range,xc+local_range,0.001)

plt.plot(x,[f(i) for i in x])
plt.plot(x,[taylor(i,x0=xc) for i in x])
plt.vlines(x=xc,ymin=0,ymax=f(xc+local_range))

plt.show()
```
