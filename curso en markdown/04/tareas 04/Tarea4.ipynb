```python
import random
y=Y
x = (X - X.mean()) / X.std()
alpha = 0.01 #Step size
iterations = 1400 #No. of iterations
m = Y.size #No. of data points
np.random.seed(123) #Set the seed
w = random.random()
b = random.random() #Pick some random values to start with

# Codigo tomado de https://www.kaggle.com/code/tentotheminus9/linear-regression-from-scratch-gradient-descent/notebook
#GRADIENT DESCENT
def gradient_descent(x, y, w,b, iterations, alpha):
    past_costs = []
    past_w = [w]
    past_b=[b]
    for i in range(iterations):

        dw1=0
        db1=0
        c=0
        for j in range (m):
            dw1=dw1+ 2*(w*x[j]+b-y[j])*x[j]
            db1=db1+2*x[j]*w+b-y[j]
            c=c+x[j]*w+b-y[j]
        dw=dw1/m
        db=db1/m
        cost=(c**2)/m
        past_costs.append(cost)
        w = w - (alpha * dw)

        b=b- (alpha*db)
        past_w.append(w)
        past_b.append(b)
        
    return past_w,past_b, past_costs

#Pass the relevant variables to the function and get the new values back...
past_w, past_b,past_costs = gradient_descent(x, Y,w,b, iterations, alpha)
theta = past_w[-1]

#Print the results...
print("Gradient Descent: {:.2f}, {:.2f}, {:.2f}".format(past_w[-1], past_b[-1], past_costs[-1]))

past_w

#Plot the cost function...
plt.title('Cost Function J')
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.plot(past_costs)
plt.show()


```

```python
df = pd.read_csv("entrenamiento_logisticRegression_withoutbias.csv")

class00 = df.iloc[:100,1:4]
class01 = df.iloc[100:,1:4]

plt.scatter(class00.iloc[:,0],class00.iloc[:,1])

plt.scatter(class01.iloc[:,0],class01.iloc[:,1])

plt.show()

import numpy as np
 
## Codigo obtenido de https://towardsdatascience.com/logistic-regression-from-scratch-in-python-ec66603592e2
#y modificado

class LogisticRegression:
    def __init__(self,x,y):      
        self.intercept = np.ones((x.shape[0], 1))  
        print(self.intercept.shape,x.shape[0])
        self.x = np.concatenate((self.intercept, x), axis=1)
        self.weight = np.zeros(self.x.shape[1])
        self.y = y
         
    #Sigmoid method
    def sigmoid(self, x, weight):
        z = np.dot(x, weight)
        return 1 / (1 + np.exp(-z))
     
    #method to calculate the Loss
    def loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
     
    #Method for calculating the gradients
    def gradient_descent(self, X, h, y):
        return np.dot(X.T, (h - y))# Aqui esta la derivada
 
     
    def fit(self, lr , iterations):
        for i in range(iterations):
            sigma = self.sigmoid(self.x, self.weight)
             
            loss = self.loss(sigma,self.y)
 
            dW = self.gradient_descent(self.x , sigma, self.y)
             
            #Updating the weights
            self.weight -= lr * dW
 
        return print('fitted successfully to data')
     
    #Method to predict the class label.
    def predict(self, x_new , treshold):
        x_new = np.concatenate((self.intercept, x_new), axis=1)
        result = self.sigmoid(x_new, self.weight)
        result = result >= treshold
        y_pred = np.zeros(result.shape[0])
        for i in range(len(y_pred)):
            if result[i] == True: 
                y_pred[i] = 1
            else:
                continue
                 
        return y_pred
    
    
#Preparing the data
x1 = np.asarray(df.iloc[:200,2:3])
y1 = np.asarray(df.iloc[:200,3])

 
#creating the class Object
regressor = LogisticRegression(x1,y1)
 
#
regressor.fit(0.1 , 5000)
 
 
y_pred = regressor.predict(x1,0.5)
 
print('accuracy -> {}'.format(sum(y_pred == y1) / y1.shape[0]))
```
